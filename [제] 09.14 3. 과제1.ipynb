```r
x=read.csv('adm.csv',header=T);
```

```r
#줄기-잎 그림
stem(x$GPA,scale=2)
```

```r
breaks=seq(from=min(x$GPA),by=(max(x$GPA)-min(x$GPA))/19, length.out=20);
hist(x$GPA,breaks=breaks,right=F, main='GPA',xlab='GPA',ylab='도수',xlim=c(0,5));
```

```r
xx=x$GPA;
xmean=mean(xx);
xmed=median(xx);#중앙값\
xsd=sd(xx);
xiqr<-quantile(xx,0.75)-quantile(xx,0.25)
#결과 출력. cat()
cat('평균=',xmean,'중앙값=',xmed,'표준편차=',xsd,'IQR=',xiqr,'\n');
```

```r
table(x$성별,x$입학전형);

#table(x$입학전형[x$성별=='남'],)
histG=function(xx,main){
    x=hist(xx,breaks=breaks,main=main,xlim=c(0,5)),x;
    
}
histG(x$GPA[x$성별=='남'],'남-GPA');
histG(x$GPA[x$성별=='여'],'여-GPA')
```

```r
Get = function (xx,name){
    xmean=mean(xx);
    xmed=median(xx);#중앙값\
    xsd=sd(xx);
    xiqr<-quantile(xx,0.75)-quantile(xx,0.25)
    #결과 출력. cat()
    cat('종류=',name,'평균=',xmean,'중앙값=',xmed,'표준편차=',xsd,'IQR=',xiqr,'\n');

}

Get(x$GPA[x$성별=='남'],'남');
Get(x$GPA[x$성별=='여'],'여');
```

```r
histG(x$GPA[x$입학전형=='학생부종합'],'학생부종합-GPA');
histG(x$GPA[x$입학전형=='학생부교과'],'학생부교과-GPA')
histG(x$GPA[x$입학전형=='정시일반'],'정시일반-GPA')
```
