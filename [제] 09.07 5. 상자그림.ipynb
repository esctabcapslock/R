```r
data7=read.csv('juice.csv',header=T);
x=data7$주스량
boxplot(x,horizontal=T,main='파괴된 주스량',col='#b8e2f2')
points(x,rep(1,19))
xfn=fivenum(x)#5분위값
text(xfn,0.65,labels=xfn)
```
