```r
#setwd('D:/data');
d1<-read.csv('res2.csv',header=F);

#as.matrix(d1); #매트릭스 변환
x<-as.vector(matrix(as.matrix(d1),ncol=1)); #매트릭스 변환개 열로 변환 -> 백터로 변환
```

```r
xmean=mean(x);
xmed=median(x);#중앙값
tab = table(x);#x의 빈도표
max(tab) #최대 빈도수
tab==max(tab) #각 빈도수가 최대인지 아닌지 확인
tab[tab==max(tab)] #최대인 값 추출
names(tab[tab==max(tab)]) #저항값만 추출
as.numeric(names(tab[tab==max(tab)])) #수로 변환
xmode=as.numeric(names(tab[tab==max(tab)])) #최빈값
#결과 출력. cat()
cat('평균=',xmean,'중앙값=',xmed,'최빈값=',xmode,'\n');
```

```r
#기하평균. 사용자 정의 함수...
gm_mean = function(a){
    prod(a)^(1/length(a));
}
gmean=gm_mean(x)#기하평균
hmean=1/mean(1/x);#기하평균
tmean=mean(x,trim=0.1) #절사평균. 10%
# 역시 결과
cat('기하평균=',gmean,'조화평균=',hmean,'절사평균=',tmean,'\n');
```

```r
#분산
xvar = var(x);
#표준편차
xsd1=sd(x); xsd2=sqrt(xvar); #두가지 방법 가능.

#범위
xrng=max(x)-min(x)
#사분위수 범위
xiqr<-quantile(x,0.75)-quantile(x,0.25)
#x의 변동계수
xcv=xsd1/mean(x)

cat('분산=',xvar,'표준편차=',xsd1,'범위=',xrng,'IQR=',xiqr,'CV=',xcv,'\n');
```
